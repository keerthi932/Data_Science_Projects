Replacing the Null Values in Columns

```python
null_cols_mean_dataset= cancer_df.groupby('Outcome', as_index= True).agg({'mean concavity': 'mean', 
                                  'mean concave points': 'mean',
                                  'concavity error': 'mean',
                                  'concave points error': 'mean',
                                  'worst concavity': 'mean',
                                  'worst concave points': 'mean'})  
null_cols_mean_dataset  
```

```python
null_cols_list= ['mean concavity', 'mean concave points', 'concavity error', 
                 'concave points error', 'worst concavity', 'worst concave points']
for nullCol in null_cols_list:
    cancer_df.loc[(cancer_df['Outcome']==0) & (cancer_df[nullCol].isnull()), nullCol]= null_cols_mean_dataset[nullCol][0]
    cancer_df.loc[(cancer_df['Outcome']==1) & (cancer_df[nullCol].isnull()), nullCol]= null_cols_mean_dataset[nullCol][1]
```

```python
cancer_df.isnull().sum()
```

Removing Outliers

```python
columns_list= ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error', 'fractal dimension error',
       'worst radius', 'worst texture', 'worst perimeter', 'worst area',
       'worst smoothness', 'worst compactness', 'worst concavity',
       'worst concave points', 'worst symmetry', 'worst fractal dimension']

fig, ax= plt.subplots(6, 5, figsize= (25, 22))
i, j= 0, 0
for col in columns_list:
    if j<5:
        sns.boxplot(cancer_df, x= col, ax= ax[i][j])
        j+=1
    else:
        i+=1
        sns.boxplot(cancer_df, x= col, ax= ax[i][0])
        j=1
```

```python
for col_name in columns_list:
    Q1= cancer_df[col_name].quantile(0.25)
    Q3= cancer_df[col_name].quantile(0.75)
    IQR= Q3 - Q1
    upper_limit= Q3 + 1.5*IQR
    lower_limit= Q1 - 1.5*IQR
    cancer_df.loc[cancer_df[col_name]> upper_limit, col_name]= upper_limit
    cancer_df.loc[cancer_df[col_name]< lower_limit, col_name]= lower_limit
```

```python
fig, ax= plt.subplots(6, 5, figsize= (25, 22))
i, j= 0, 0
for col in columns_list:
    if j<5:
        sns.boxplot(cancer_df, x= col, ax= ax[i][j])
        j+=1
    else:
        i+=1
        sns.boxplot(cancer_df, x= col, ax= ax[i][0])
        j=1
```

Training And Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, random_state= 0, test_size= 0.2, stratify= y)
```

```python
print('x train shape:', x_train.shape)
print('x test shape:', x_test.shape)
print('y train shape:', y_train.shape)
print('y test shape:', y_test.shape)
```

**IMPLEMENTING ML MODELS**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

Logistic Regression

```python
lg= LogisticRegression()
lg.fit(x_train, y_train)
y_predict_lg= lg.predict(x_test)
accuracy_score_lg= accuracy_score(y_test, y_predict_lg)
print('training accuracy score:', accuracy_score(y_train, lg.predict(x_train)))
print('testing accuracy score:', accuracy_score_lg)
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict_lg), annot= True)
```

```python
print(classification_report(y_test, y_predict_lg))
```

SVC

```python
svc= SVC(probability= True)
parameters= {
    'C': [0.0001, 0.001, 0.01, 0.1, 1.0],
    'kernel': ['linear', 'poly', 'sigmoid', 'rbf'],
    'degree': [3, 5, 7, 10],
    'gamma': [0.001, 0.01, 0.1, 1.0]
}
gridSearch_svc= GridSearchCV(svc, parameters)
gridSearch_svc.fit(x_train, y_train)
```

```python
gridSearch_svc.best_params_
```

```python
gridSearch_svc.best_score_
```

```python
svc= SVC(C= 0.0001, degree= 7, gamma= 0.01, kernel= 'poly')
svc.fit(x_train, y_train)
y_predict_svc= svc.predict(x_test)
accuracy_score_svc= accuracy_score(y_test, y_predict_svc)
print('training accuracy score:', accuracy_score(y_train, svc.predict(x_train)))
print('testing accuracy score:', accuracy_score_svc)
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict_svc), annot= True)
```

```python
print(classification_report(y_test, y_predict_svc))
```
